```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
car = pd.read_csv('carprices.csv')
car
```

```python
x = car.drop(columns=['Sell Price($)'])
y = car['Sell Price($)']
```

```python
# %matplotlib inline
# x =plt.xlabel(['Mileage','Age(yrs)'])
# y =plt.ylabel('Sell Price($)')
# plt.scatter(car.x,car.y, color='blue', marker='+')
```

```python
dummies = pd.get_dummies(car['Car Model'])
dummies
```

```python
merg = pd.concat([car,dummies],axis='columns')
merg
```

```python
final = merg.drop(['Car Model', 'Mercedez Benz C class'],axis='columns')
final
```

```python
x = final.drop(columns=['Sell Price($)'])
y = final['Sell Price($)']
x 
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x,y)
```
